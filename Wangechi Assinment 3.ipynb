```python
#1. Show a breakdown of distance from home by job role and attrition.
Breakdown = IBM.groupby(["JobRole","Attrition"])["DistanceFromHome"].mean().unstack().fillna(0)
Breakdown
```

```python
Breakdown.plot(kind='bar', stacked=True)
```

```python
#2. Compare average monthly income by education and attrition.
Compare=IBM.groupby(["Education","Attrition"])["MonthlyIncome"].mean().unstack().fillna(0)
Compare
```
